**Wörter verschlüsseln**  
Das Modul `hashlib` bietet verschiedene Verschlüsselungsmöglichkeiten an,  
von denen allerdings keines wirklich für Kennwörter verwndet wird!  
Wir spielen ja nur ...

```python
hashlib.sha256(wort.strip().encode()).hexdigest()
```

*Erklärung*  

* `sha256` ist das Verfahren, das wird sehr wohl zum Verschlüsseln verwendet
  aber in anderem Zusammenhang!
* `strip()` entfernt Leerzeichen und Zeilenumbrüche an den Worträndern (siehe oben)
* `encode` wandelt das Wort in reine Bytes um. Buchstaben wie "ö" werden
  dabei passend umgewandelt - siehe Codefenster unten!
* `hexdigest` liefert den codierten Text als Zeichenkette

````python
# Encode und seine Wirkung  
text_1 = "Regenwurm"
print(text_1.encode())   # Im Ergebnis steht b"..." für "Byte-String"
text_2 = "Regenwürmer"
print(text_2.encode())   # ü wird zu #c3bc
````

## Wie zerlegt man Zeichenketten mit Trennzeichen?

Man spricht dabei vom *splitten* eines Strings und entsprechend heißt der   
Befehl auch `split`. Es gibt zwei sinnvolle Varianten:

```python
text = "Das;ist;ein;Test"
wort1, wort2, wort3, wort4 = text.split(";")
woerter = text.split(";")
```

In der zweiten Version bekommst du eine Liste mit dem Namen `woerter`  
deren Inhalt du mit `woerter[0]` bis `woerter[3]`  auslesen kannst.  
Das ist sinnvoll, wenn du vorher nicht weißt, wie viele *Spalten*  
entstehen werden.

**Tipp für den Krimi**

## Listen, Mengen und Dictionaries

Wir kennen in Python verschiedene Möglichkeiten, um größere Mengen  
an Daten so abzuspeichern, dass wir wieder gut darauf zugreifen können.  
Jede Variante hat ihre speziellen Vor-und Nachteile.

### Listen

Hier werden Daten hintereinander gepackt und wir können über ihre  
Nummer in der Reihenfolge zugreifen:

```python
namen = []  # noch ist die Liste leer
namen.append("Maier")  # jetzt ist es ["Maier"]
namen.append("Huber")  # jetzt ist es ["Maier", "Huber"]
...
# Zugriff
namen[0]  # steht für "Maier"
namen[1]  # steht für "Huber"
```

### Mengen (engl. set)


In einer Menge gibt es keine Reihenfolge und keine doppelten Einträge.  
*Set*s sind praktisch zum Zählen, wenn man die Doppelten rausfiltern will!

```python
namen = set()
namen.add("Maier")  # enthält Maier
namen.add("Huber")  # enthält Maier und Huber
namen.add("Maier")  # enthält Maier und Huber

# Zugriff
# Ist nicht so einfach, weil es ja keine Reihenfolge gibt
# Vorher in Liste umwandeln
namenliste = list(namen)
print(namenliste[0])
print(namenliste[1])
...
```

**Tipp**  
Der 'len'-Befehl geht hier auch um die Länge der Menge zu ermitteln
