### Wie entpacke ich eine Datei?

Du musst wissen, wo die Datei gespeichert ist und wie sie heißt. 
Für dieses Beispiel liegt eine Datei mit dem Namen `personen.zip`  
im Ordner `training` für dich bereit.

**Entpacken**  
Da Python auch auf sehr kleinen Computern funktionieren muss, darf Python  
nicht besonders groß sein. Um hier flexibel zu sein, besteht Python  
aus vielen Modulen, die nur bei Bedarf **importiert** werden.  

Zum Entpacken von Dateien ist das Modul **shutil** hilfreich.

```python
import shutil
```

Wenn du den Speicherort kennst (Ordner `training`) und auch den   
Namen der Datei (`personen.zip`) dann kannst du sie mit folgendem   
Befehl entpacken:

```python
shutil.unpack_archive("training/personen.zip", "training")
```

Die entpackte Datei liegt dann im Ordner `training`, weil du das als  
als Ziel angegeben hast. Du kannst nun auch einen Blick in diese Datei  
werfen, indem du die entpackte Version `personen.csv` anklickst.

if "llo" in text:  # wahr

# Wie zerelgt man Zeichenketten mit Trennzeichen?

Man spricht dabei vom *splitten* eines Strings und entsprechend heißt der   
Befehl auch `split`. Es gibt zwei sinnvolle Varianten:

```python
text = "Das;ist;ein;Test"
wort1, wort2, wort3, wort4 = text.split(";")
woerter = text.split(";")
```

In der zweiten Version bekommst du eine Liste mit dem Namen `woerter`  
auf die du über `woerter[0]` bis `woerter[3]`  zugreifen kannst.  
Das ist sinnvoll, wenn man vorher nicht weiß, wie viele *Spalten* im   
Text enthalten sind.  

**Tipp**  
> Auch der Tabulator ist ein Trennzeichen. Den musst du beim Splitten  
> aber als "\t" schreiben!



# Listen, Mengen und Dictionaries

Wir kennen in Python verschiedene Möglichkeiten, um größere Mengen  
an Daten so abzuspeichern, dass wir wieder gut darauf zugreifen 
können. Jede Variante hat ihre speziellen Vor-und Nachteile.

## Listen

Hier werden Daten hintereinander gepackt und wir können über ihre  
Nummer in der Reihenfolge zugreifen:

```python
namen = []  # noch ist die Liste leer
namen.append("Maier")  # jetzt ist es ["Maier"]
namen.append("Huber")  # jetzt ist es ["Maier", "Huber"]
...
# Zugriff
namen[0]  # steht für "Maier"
namen[1]  # steht für "Huber"
```

## Mengen (engl. set)
In einer Menge gibt es keine Reihenfolge und keine doppelten Einträge.
*Set*s sind praktisch zum Zählen, wenn man die Doppelten rausfiltern will!

```python
namen = set()
namen.add("Maier")  # enthält Maier
namen.add("Huber")  # enthält Maier und Huber
namen.add("Maier")  # enthält Maier und Huber

# Zugriff
# Ist nicht so einfach, weil es ja keine Reihenfolge gibt
# Vorher in Liste umwandeln
namenliste = list(namen)
print(namenliste[0])
print(namenliste[1])
...
```

**Tipp**  
Der 'len'-Befehl geht hier auch um die Länge der Menge zu ermitteln
